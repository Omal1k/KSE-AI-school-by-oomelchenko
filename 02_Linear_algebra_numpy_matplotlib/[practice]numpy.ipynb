***Функція `np.arange` схожа до `range`. Аргуенти можуть бути з не цілими числами. Треба запобігати ситуації, коли (кінець-початок) / шаг - ціле число, бо в цьому випадку включення останнього елменту залежить від помилки округлення. Краще, щоб кінець діапазонк був десь посередині кроку.***

```python
a = np.arange(1, 16, 4)
b = np.arange(5., 21, 2)
c = np.arange(1, 10)
d = np.arange(5)
print(a)
print(b)
print(c)
print(d)
```

***Послідонвість чисел з постійним кроком можливо також створювати функцією `linspace`. Початок і кінець діапазону включається; останній аргумент -- число крапок.***

```python
a = np.linspace(1, 15, 2)
b = np.linspace(5, 12, 10)
print(a)
print(b)
```

**Завдання 3:**  
- Створити масив чисел від $-4\pi$  до $4\pi $, кі-ть точок 100
- порахувати суму поелментних квадратів синуса і косинуса для даного масиву.  
- За допомогою ``np.all`` перевірити, що всі елменти дорівюють одному.

```python
# вирішення

x = np.linspace(-4*np.pi, 4*np.pi, 100)

np.all((np.sin(x)**2 + np.cos(x)**2).round() == 1)
```

***Двовимірні масиви, які залежать тільки від одного індексу: $x_{ij}=u_j$, $y_{ij}=v_i$***

# 5. Лінійна алгебра
